```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "listings2.csv"
data_df = pd.read_csv(csv_file)
```

```python
data_df.head()
```

```python
data_df['review_scores_rating'] = data_df['review_scores_rating'].astype(int)
```

```python
data_df['neighbourhood_cleansed'] = data_df['neighbourhood_cleansed'].astype(str)
```

```python
data_df['price'] = data_df['price'].astype(int)
```

```python
data_df['monthly_price'] = data_df['monthly_price'].astype(int)
```

```python
# reviews_per_month,  accomodates

data_df['reviews_per_month'] = data_df['reviews_per_month'].astype(int)
```

```python
data_2= data_df.copy()
```

```python
database_path = "project2"
engine = create_engine(f"sqlite:///project2.sqlite")
```

```python
engine.table_names() 
```

```python
data_2.to_sql(name='listings', con=engine, if_exists='append', index=True) 
```
